```python
import pandas as pd 
import statsmodels as sm
import numpy as np
import matplotlib.pyplot as plt
```

```python
state = pd.read_csv('../Csv/state.csv')

```

```python
state['Murder.Rate'].quantile([0.05,0.25,0.5,0.75,0.95])
```

```python
#Codigo para crear un diagrama de caja
ax = (state ['Population']/1_000_000).plot.box()
ax.set_ylabel('Population (millones)')
```

```python
#Codigo para crear una segmentacion por rango de la poblacion
binnedPopulation = pd.cut(state['Population'],7)
binnedPopulation.value_counts()
```

```python
#Codigo para crear un histograma con pandas y matplotlib

ax = (state['Population']/1_000_000).plot.hist(figsize=(4,4))
ax.set_xlabel('Population (millones)')
```

```python
#Crear un digrama de densidad con panda y python
ax = state['Murder.Rate'].plot.density()
ax.set_xlabel('Tasa homicidios')
```
