```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn as sk
import scipy.stats as stats
```

```python
#Generar un diagrama QQ con probplot de scipy

#Importamos la libreria scipy.stats
import scipy.stats as stats

#Generamos un diagrama QQ con probplot de scipy
fig, ax = plt.subplots(figsize=(4,4))

#Generamos una muestra de 100 datos de una distribuci√≥n normal
loans_sample = stats.norm.rvs(size =100)

#Generamos el diagrama QQ
stats.probplot(loans_sample, plot=ax)
```

```python
#Crear una distribucion de cola larga en Python
#Importamos la libreia numpy
import numpy as np

#Leemos el archivo csv con los datos
sp500 = pd.read_csv('../Csv/sp500_data.csv')

#Seleccionamos la columna de los datos de la empresa Netflix
nflx = sp500.NFLX
#Calculamos los rendimientos logaritmicos
nflx= np.diff(np.log(nflx[nflx>0]))

#Generamos un diagrama QQ con probplot de scipy
fig,ax = plt.subplots(figsize=(4,4))

#Generamos el diagrama QQ
stats.probplot(nflx, plot = ax)
```
