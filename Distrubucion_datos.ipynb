```python
import pandas as pd 
import statsmodels as sm
import numpy as np
import matplotlib.pyplot as plt
```

```python
state = pd.read_csv('Csv/state.csv')

```

```python
state['Tasa'].quantile([0.05,0.25,0.5,0.75,0.95])
```

```python
#Codigo para crear un diagrama de caja
ax = (state ['Poblacion']/1_000_000).plot.box()
ax.set_ylabel('Poblacion (millones)')
```

```python
#Codigo para crear una segmentacion por rango de la poblacion
binnedPopulation = pd.cut(state['Poblacion'],7)
binnedPopulation.value_counts()
```

```python
#Codigo para crear un histograma con pandas y matplotlib

ax = (state['Poblacion']/1_000_000).plot.hist(figsize=(4,4))
ax.set_xlabel('Poblacion (millones)')
```

```python
#Crear un digrama de densidad con panda y python
ax = state['Tasa'].plot.density()
ax.set_xlabel('Tasa homicidios')
```
